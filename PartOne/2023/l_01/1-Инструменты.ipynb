# Jupyter Notebook
Почти весь код мы будем писать в приложении Jupyter Notebook. 
C помощью Jupyter Notebook можно создавать интерактивные ноутбуки (с английского notebook — тетрадь) с кодом на языке Python и текстовыми элементами. Jupyter работает с файлами формата .ipynb. Результатом работы в ноутбуке является код, который презентует сам себя с помощью текстовых комментариев.

Код в ноутбуке разделён на блоки, которые можно выполнять в произвольном порядке. После выполнения блока результат его работы сохраняется в оперативной памяти Jupyter Notebook. Это возможно благодаря тому, что в начале работы Jupyter Notebook запускает ядро (kernel), которое производит все операции. Ядро можно перезагружать, но тогда все сохранённые переменные сотрутся.

Для прохождения курса вам понадобится несколько библиотек для работы с данными. Поэтому мы рекомендуем устанавливать Jupyter как часть пакета Anaconda, куда помимо Jupyter добавлены самые распространённые библиотеки, которые в том числе понадобятся вам в этом курсе. 
Установить Anaconda можно по ссылке (https://www.anaconda.com/download).

Также даём ссылку (https://docs.anaconda.com/free/anaconda/getting-started) на quick-start с Jupyter Notebook.

# Google Colab
Google Colab — это сервис компании Google, который обеспечивает возможность запускать Jupyter Notebook-и онлайн на серверах Google. Часто это оказывается удобнее, поскольку нет необходимости устанавливать Jupyter на свой компьютер. В начале курса вы можете пользоваться этим вариантом, хотя мы рекомендуем всё же установить Anaconda. Например, автору этого текста интерфейс Jupyter Notebook нравится гораздо больше, чем интерфейс Google Colab.

К каждому видео, в котором мы обсуждаем код, будет приложена ссылка на ноутбук в Google Colab. 
Сейчас мы даём ссылку на обучающий ноутбук (https://colab.research.google.com/notebooks/intro.ipynb) с базовой информацией о Colab. 
Пожалуйста, пройдите по ней сейчас, иначе дальнейший текст будет непонятен.

Чтобы самостоятельно исполнить этот код и иметь возможность его редактировать, вам необходимо скопировать ноутбук на свой Google Drive (на странице с ноутбуком Файл -> создать копию на диске, скаченный файл: **2-Welcome_To_Colaboratory.ipynb**).


Пожалуйста, ознакомьтесь со вступительным ноутбуком и выполните инструкции.

Как правило, нам придётся загружать в Google Colab локальные данные с диска. В этом случае вам понадобится скопировать данные в свою директорию на Google Drive и обеспечить сервису Google Colab доступ к вашему Drive. Образец кода для предоставления доступа будет в одном из первых занятий.

Ещё одна приятная черта Google Colab — это возможность использовать GPU-память, установленную на серверах Colab. Обычно персональные компьютеры не снабжены GPU, подходящими для быстрого обучения нейронных сетей. Поэтому Colab окажется незаменимым помощником для быстрого выполнения простых заданий с нейронками.

Иногда для особо сложных задач даже мощностей Colab оказывается недостаточно. Но об этом мы поговорим тогда, когда такие задания появятся в нашем курсе:)

# Командная строка
Для комфортного прохождения курса вам часто нужна будет командная строка. Впрочем, курс можно проходить и без неё (например, если всё делать на Colab).

Jupyter Notebook имеет важное преимущество: из него можно исполнять команды, как бы притворяясь, что вы находитесь в командной строке. Мы прикладываем обучающий ноутбук (https://colab.research.google.com/drive/1P0RE43Ih9J9TmXkrpgfNwjFgsRIF3BPe, скаченный файл: **3-bash.ipynb**), в котором основной функционал командной строки описан.
