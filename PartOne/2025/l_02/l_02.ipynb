### Jupyter Notebook
Основная часть работы в этом курсе будет проходить в среде Jupyter Notebook. Это мощное приложение, которое позволяет создавать интерактивные "тетради" (notebooks) с кодом на Python, текстовыми пояснениями, графиками и другими элементами. Jupyter Notebook работает с файлами формата .ipynb и отлично подходит для анализа данных, визуализации и запуска небольших экспериментов.

Как устроен Jupyter Notebook?
Код в ноутбуке разделён на блоки (ячейки), которые можно выполнять в любом порядке. После выполнения блока результат (например, вывод данных или график) сохраняется в оперативной памяти Jupyter Notebook. Это возможно благодаря ядру (kernel), которое запускается при старте Jupyter Notebook и выполняет все вычисления. Если ядро перезагрузить, все сохранённые переменные и данные будут удалены.

Что нужно для начала работы?
Для комфортной работы в Jupyter Notebook мы рекомендуем установить пакет Anaconda. Он включает в себя не только Jupyter, но и самые популярные библиотеки для работы с данными. Это избавит вас от необходимости устанавливать их вручную. Скачать Anaconda можно по [этой ссылке](https://www.anaconda.com/products/individual).

Быстрый старт с Jupyter Notebook
Если вы новичок в Jupyter, рекомендуем ознакомиться с [quick-start](https://docs.anaconda.com/anaconda/user-guide/getting-started) гайдом, который поможет вам быстро освоить основные функции и начать работу.

### Kaggle
Здесь мы с вами рассмотрим альтернативу Google Colab для работы c Jupyter Notebooks.

[Kaggle](https://www.kaggle.com) — это платформа для соревнований по анализу данных и машинному обучению, а также удобная среда для работы с Jupyter Notebook. Kaggle предоставляет доступ к мощным вычислительным ресурсам бесплатно, но с ограничениями.

В начале обучения можно использовать Google Colab или локальное окружение, но Kaggle — отличный выбор для работы с готовыми датасетами и тестирования моделей в реальных условиях соревнований. В этом руководстве мы рассмотрим его преимущества и основные шаги для начала работы.

Почему стоит использовать Kaggle?
1. Готовая среда с Jupyter Notebook
Kaggle предоставляет встроенную среду Jupyter Notebook, где можно писать и запускать код без необходимости настраивать локальное окружение.

2. Бесплатный доступ к GPU/TPU
Пользователи Kaggle получают ограниченный, но бесплатный доступ к вычислительным ресурсам, включая GPU (NVIDIA Tesla P100) и TPU. Это полезно для глубокого обучения и работы с большими моделями.

3. Доступ к огромному количеству датасетов
На Kaggle размещено множество датасетов в открытом доступе, что позволяет быстро начать работу с реальными данными без необходимости загружать их вручную.

4. Сообщество и соревнования
Kaggle известен своими соревнованиями по анализу данных, где можно решать реальные бизнес-задачи, улучшать навыки в Data Science и даже выигрывать денежные призы. Также можно изучать решения других участников и обучаться на их примерах.

5. Возможность сохранения и публикации моделей
Пользователи могут сохранять свои модели, делиться ими и использовать репозитории других специалистов в области машинного обучения.

Как начать работу с Kaggle?
1. Создать аккаунт на Kaggle
Для начала работы необходимо зарегистрироваться на сайте [Kaggle](https://www.kaggle.com) и создать профиль.

1) Переходишь по ссылке: https://www.kaggle.com/contact#/account/activate/phone 

2) Нажимаешь синюю кнопку "contact support" 

3) Заполняешь страну "Russia", номер телефона 7XXXYYYYYYY (без плюсов, тире и т.д., только 7 и цифры) 

4) В тексте сообщения пишешь: Hi! I can't verify number 7XXXYYYYYYY, got the message "This phone number can't be verified". Please help with this issue. Далее на почту (обычно в течение 1-2 дней) придет письмо, что можно верифицировать телефон.

2. Открыть Kaggle Notebooks
Kaggle предлагает удобную среду для запуска Jupyter Notebook. Для этого:

Перейдите в раздел "Code".

Нажмите "New Notebook".

Выберите нужные ресурсы (например, GPU/TPU, если требуется).

3. Загрузить или выбрать датасет
Можно использовать датасеты из Kaggle или загружать свои. Для этого:

Перейдите в раздел "Datasets"

Подключите его в Notebook с помощью !kaggle datasets download или через встроенные инструменты.

Настроить окружение можно с помощью стандартных Python-библиотек (pandas, numpy, sklearn, tensorflow и др.), которые уже предустановлены в Kaggle Notebook.

5. Участие в соревнованиях
Можно участвовать в соревнованиях Kaggle:

Перейдите в раздел "Competitions".

Выберите активное соревнование.

### Yandex Datasphere
Yandex DataSphere — это облачный сервис компании Яндекс, который обеспечивает удобную среду для работы с Jupyter Notebook и позволяет запускать вычисления на мощных серверах.

В начале курса вы можете использовать Google Colab или Kaggle, но для более сложных задач DataSphere — отличный выбор. В этом уроке мы рассмотрим его плюсы и как начать работу с ним.

Почему стоит использовать DataSphere?
DataSphere — это платформа, которая объединяет удобство Jupyter Notebook с мощностью облачных вычислений. Основные преимущества:

* Гибкое управление ресурсами. Можно выбирать мощность виртуальной машины (CPU/GPU), а также автоматически останавливать вычисления при простое. Например, если вы обучаете модель, а затем оставили ноутбук работать, DataSphere отключит вычисления, чтобы не расходовать бюджет.

* Единое рабочее пространство. Весь код, данные и вычислительные ресурсы хранятся в одном месте, без необходимости вручную загружать файлы. Это особенно удобно при работе с большими датасетами, например, при анализе медицинских изображений или обработке данных финансовых рынков.

* Совместная работа. Можно приглашать коллег, делиться кодом и ресурсами в рамках проектов. Команды Data Science часто работают с большими моделями, и DataSphere позволяет эффективно распределять нагрузку между участниками.

* Подключение к S3-хранилищам. Храните большие датасеты в облаке и подключайте их без скачивания на локальный диск. Например, если вы анализируете данные со спутниковых снимков, S3-хранилища позволяют загружать только нужные фрагменты. [Подробнее](https://yandex.cloud/ru/docs/storage/tutorials/s3-to-datasphere)

* Удалённый запуск скриптов. DataSphere поддерживает DataSphere Jobs, позволяя запускать обучение в облаке, даже если ноутбук выключен. Это особенно полезно при долгих экспериментах с моделями. [Подробнее](https://yandex.cloud/ru/docs/datasphere/concepts/jobs)

* Встроенный хостинг моделей. После обучения можно развернуть модель и настроить её API для использования в продакшене, например, для чат-ботов, рекомендательных систем или распознавания речи. [Подробнее](https://yandex.cloud/ru/docs/datasphere/operations/deploy/node-create)

Как начать работу с DataSphere?
1. Создать аккаунт в Yandex Cloud. Нужно [зарегистрироваться](https://yandex.cloud/ru/docs/datasphere/quickstart) и активировать бесплатный тестовый период.

2. Создать проект в DataSphere. Это ваше рабочее пространство, в котором будут храниться ноутбуки, данные и вычислительные ресурсы.

3. Запустить JupyterLab. Интерфейс полностью аналогичен Jupyter Notebook, поэтому переход с Colab не вызовет трудностей.

4. Настроить ресурсы. Вы можете выбрать минимально необходимую конфигурацию и при необходимости увеличивать мощность.

5. Загрузить данные. Можно подключить облачное S3-хранилище или загружать файлы напрямую в проект.

6. Начать обучение моделей. После настройки среды можно запускать код так же, как в Colab, но без ограничений на время выполнения.



### Yandex Cloud AI Studio
Yandex Cloud AI Studio - еще один полезный инструмент, который может пригодится в курсе. Это платформа для разработки и развертывания AI-решений. Она объединяет сервисы для работы с генеративными моделями, синтезом и распознаванием речи, компьютерным зрением и машинным обучением.

1. YandexGPT API — генерация текстов и чат-боты
Сервис для работы с генеративными моделями текста. Позволяет создавать чат-ботов, писать статьи, формировать описания товаров и отвечать на вопросы на естественном языке.

2. SpeechKit — распознавание и синтез речи
Технология преобразования речи в текст и наоборот. Поддерживает разные языки, акценты и может использоваться для голосовых помощников, диктовки текста и озвучивания контента. С его помощью легко создать голосового ассистента или автоматически транскрибировать подкастов.

3. AI Playground — тестирование AI-моделей
Интерактивная среда для быстрого тестирования возможностей AI-моделей без необходимости писать код. Можно вводить текстовые запросы и сразу видеть ответы нейросети.

4. ML SDK — инструменты для работы с моделями
Yandex Cloud ML SDK — это библиотека для Python, упрощающая работу с фундаментальными моделями и интеграцию с сервисами Yandex Cloud. Она предоставляет удобные API для взаимодействия с AI Studio и поддерживает как синхронный, так и асинхронный вызовы через gRPC.

Для чего можно использовать ML SDK:

* Генерация текста с использованием всех доступных моделей.

* Работа с эмбеддингами для обработки и поиска текстов.

* Создание и дообучение классификаторов на основе YandexGPT.

* Разработка AI-ассистентов с индивидуальной логикой.

* Генерация изображений с помощью YandexART.

* Интеграция с LangChain для построения сложных AI-пайплайнов.

ML SDK позволяет быстро разрабатывать и развертывать AI-решения, не углубляясь в детали низкоуровневых API.
