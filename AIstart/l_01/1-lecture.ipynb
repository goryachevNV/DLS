# Чем занимаются специалисты по Data Science
В этом видео мы поговорим о том, какие специалисты работают в области Data Science и чем они занимаются. Все профессии в этой области так или иначе связаны с различными этапами работы с данными: сбор, обработка, хранение и анализ данных данных, а также построение моделей машинного обучения на основе данных.
